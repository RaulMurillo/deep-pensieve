```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
DIR = '../src/TensorFlow/train_results/'
```

### MNIST

```python
SUBDIR = DIR+'lenet5/mnist/'

# , usecols=['val_loss', 'val_acc'])
float32_hist = pd.read_csv(SUBDIR+'float32.csv', header='infer')
# , usecols=['val_loss', 'val_acc'])
posit32_hist = pd.read_csv(SUBDIR+'posit32.csv', header='infer')
# , usecols=['val_loss', 'val_acc'])
posit16_hist = pd.read_csv(SUBDIR+'posit16.csv', header='infer')

float32_hist.index += 1  # Start at 1, after first epoch
posit32_hist.index += 1
posit16_hist.index += 1

float32_hist.head()
```

```python
# FSIZE = (8, 6)

fig, axis = plt.subplots(nrows=2, ncols=1,
                         sharex=True,
                         #  figsize=FSIZE
                         )
colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd',
          '#8c564b', '#e377c2', '#7f7f7f', '#bcbd22', '#17becf']


for i, df in enumerate([float32_hist, posit32_hist, posit16_hist]):
    # Plot
    df['loss'].plot(kind='line',
                    xlabel='Epoch',
                    ylabel='Loss',
                    ax=axis[0],
                    color=colors[i],
                    )
    df['val_acc'].plot(kind='line',
                       xlabel='Epoch',
                       ylabel='Test Accuracy',
                       ax=axis[1],
                       color=colors[i],
                       )

# Some adjustments
axis[0].grid(zorder=0)
axis[1].grid(zorder=0)
# axis[0].set_ylim(bottom=0.0)
axis[1].set_ylim(top=0.995)
fig.tight_layout()

# Create a legend object for all subplots
handles, _ = axis[0].get_legend_handles_labels()
labels = ['Float32', 'Posit32', 'Posit16']
fig.legend(handles, labels, title='Format',
           loc='upper center', bbox_to_anchor=(0.5, 1.1),
           ncol=len(labels))

# save figure
plt.savefig(SUBDIR+'mnist_training.pdf', bbox_inches='tight')

plt.show()
```
